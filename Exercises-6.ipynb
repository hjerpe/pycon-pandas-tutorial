### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.

```python
sales2 = sales2[sales2.Title.notnull()]
sales2
```

```python
sales1['Title'] = sales1['Book title']

currency_values = sales2.Title.str.extract(
    '\w*\((\w*)\)')

sales_or_reader_indices = sales2.Title.str.contains(
    'Sales\ report\ for\ Q4|E\-Book\ Reader \w*\ Store')

sales2_cleaned = sales2.copy()
sales2_cleaned['Currency'] = currency_values.fillna(
    method='bfill')

sales2_cleaned.loc[
    sales_or_reader_indices, 'Currency'] = pd.NaT
sales2_cleaned = sales2_cleaned.fillna('')
sales2_cleaned
```

```python
unit_indices = sales2_cleaned['Units sold'] != ''

sales2_cleaned.loc[unit_indices, 'Total revenue'] = \
    sales2_cleaned.loc[unit_indices, 'List price'] *\
    sales2_cleaned.loc[unit_indices, 'Units sold']
    
revenues = sales2_cleaned[
    ['Total revenue', 'Currency', 'Title']].\
    fillna('')
revenues = revenues[revenues['Total revenue'] != '']

revenue_by_currency = revenues.groupby(
    ['Currency', 'Title']).sum()
revenue_by_currency
#revenue_by_currency.plot(kind='bar')
```
